## preprocessing 

```python
## separtion du jeud donnee 
Dtrain, Dtest = train_test_split(fraud, test_size=0.2)
X_train , y_train = Dtrain.drop(columns=["Class"]), Dtrain.Class
X_test , y_test = Dtest.drop(columns=["Class"]), Dtest.Class
```

```python
# feature selection 
selector = VarianceThreshold(threshold=0.5)
selector.fit(X_train)
X_train_select = selector.transform(X_train)
```

```python
# normaliser le jeu de donn√©e 
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train_select)

# Utilisation de la class PolynomialFeatures 
poly = PolynomialFeatures(degree=2)
poly.fit(X_train_scaler)
X_train_scaler =  poly.transform(X_train_scaler)
```

```python
#### Creation du model 
model = LogisticRegression()
model.fit(X_train_scaler, y_train)
model.score(X_train_scaler, y_train)
```

#### Erreur de generalisation du model 

```python
#feature selction 

X_test_select = selector.transform(X_test)
# normalisation du jeu de test 
X_test_scaler = scaler .transform(X_test_select)

# PolynomialFeatures dans le jeu de test 

X_test_scaler = poly.transform(X_test_scaler)
```

```python
model.score(X_test_scaler, y_test)
```

#### Utilisation des Pipline 

```python
iris = sns.load_dataset("iris")
X, y = iris.drop(columns=["species"], axis=1),  iris.species 
```

```python
# Normalisation 
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X, y)

# polynomialFeatures 
poly = PolynomialFeatures(degree=1)
X_scaler = poly.fit_transform(X_scaler)
```

```python
model_L = LogisticRegression()
model_L.fit(X_scaler, y)
model_L.score(X_scaler, y)
```

```python
model = make_pipeline(StandardScaler(), PolynomialFeatures(),
                                        VarianceThreshold(), 
                                        LogisticRegression()
                                        )
                                        

params_dico = {'variancethreshold__threshold':np.linspace(0,1), 
    'polynomialfeatures__degree': [1,2,3]
}
```

```python
grid = GridSearchCV(model, param_grid=params_dico)
grid.fit(X, y)
```

```python
model = grid.best_estimator_
```

```python
model.score(X, y)
```
